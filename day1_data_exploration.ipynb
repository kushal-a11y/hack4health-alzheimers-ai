```python
df = df.drop(columns=['Unnamed: 0', 'id'], errors="ignore")
df = df.dropna()
```

```python
df['age'].head()
```

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
```

```python
X = df.drop('cardio', axis=1)
y = df['cardio']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Trainign logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print('Random Forest: ')
y_pred = rf.predict(X_test)
print('Accuracy Score:', accuracy_score(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
print('Logistic Regression: ')
y_pred = lr.predict(X_test)
print('Accuracy Score:', accuracy_score(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

features = df.drop('cardio', axis=1).columns
importances = rf.feature_importances_

feature_importance = pd.DataFrame({
    'feature': features,
    'importance': importances
}).sort_values(by='importance', ascending=False)
feature_importance

plt.figure()
plt.barh(feature_importance['feature'], feature_importance['importance'])
plt.gca().invert_yaxis()
plt.title("Feature Importance for Cardiovascular Risk Prediction")
plt.show()
```

```python
#Logistic Regression Coefficients (More Technical Depth)
#  Meaning:
# positive value  increases disease risk
# negative value  protective effect

coef_df = pd.DataFrame({
    'feature': features,
    'coefficient': lr.coef_[0]
}).sort_values(by='coefficient', key=abs, ascending=False)

coef_df.head(10)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
plt.show()
```

# Advanced Feature Interaction
creation of medically meaningful interaction features
that actually boost performance.


1.   Pulse_pressure = ap_hi(Action potential) - ap_lo
2.   Metabloic_risk = cholesterol + glucose(gluc)


1.   Lifestyle_risk = smoking(smoke) + alcoholic(alco) - active

Retraining now to see differences in Accuracy, Recall(Missing sick cases), F1

```python
mdf = df.copy() #modified df
mdf.head()
```

```python
mdf['metabolic_risk'] = mdf['smoke'] + mdf['alco']
mdf['lifestyle_risk'] = mdf['smoke'] + mdf['alco'] - mdf['active']
X = mdf.drop('cardio', axis=1)
y = mdf['cardio']
#
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mrf = RandomForestClassifier(
    n_estimators=400,
    max_depth=14,
    random_state=42,
    n_jobs=-1)
mrf.fit(X_train, y_train)

print('Random Forest: ')
y_pred = mrf.predict(X_test)
print('Accuracy Score:', accuracy_score(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
mlr = LogisticRegression(max_iter=3000)
mlr.fit(X_train, y_train)

print('Logistic regression: ')
y_pred = mlr.predict(X_test)
print('Accuracy Score:', accuracy_score(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
features = mdf.drop('cardio', axis=1).columns
importance = mrf.feature_importances_
fimp = pd.DataFrame({
    'feature': features,
    'importance': importance})
fimp.sort_values(by='importance', ascending=False)
```

```python
importance = mlr.coef_[0]
fimp = pd.DataFrame({
    'feature': features,
    'importance': importance})
fimp.sort_values(by='importance', ascending=False)
```
